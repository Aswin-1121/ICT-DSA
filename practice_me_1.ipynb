# Bar Graph

```python
plt.bar(dat['Classification'],dat['PL'],color ='green', width =.8)
plt.title("classification based on PL")
plt.xlabel("Types")
plt.ylabel("Petal Length(PL)");

# Here iris virginica has higher petal length
```

```python
plt.bar(dat['Classification'],dat['PW'],color ='yellow', width =.8)
plt.title("classification based on PW")
plt.xlabel("Types")
plt.ylabel("Petal Width(PW)");

# Here iris virginica has higher petal width
```

```python
plt.bar(dat['Classification'],dat['SW'],color ='purple', width =.8)
plt.title("classification based on SW")
plt.xlabel("Types")
plt.ylabel("Sepal Width(SW)");

# Here iris setosa has higher sepal width
```

```python
plt.bar(dat['Classification'],dat['SL'],color ='red', width =.8)
plt.title("classification based on SL")
plt.xlabel("Types")
plt.ylabel("Sepal Length(SL)");

# Here iris virginica has higher sepal length
```

# Scatter Plot

```python
sns.scatterplot(x='SL', y='SW', hue='Classification', data=dat)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()

# From below graph,

# 1.iris Setosa has smaller sepal lengths but larger sepal widths.
# 2.Versicolor  lies in the middle of the other two in terms of sepal length and width
# 3.Virginica has larger sepal lengths but smaller sepal widths.
```

```python
sns.scatterplot(x='PL', y='PW', hue='Classification', data=dat)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()

# From below graph,

# 1.Setosa has smaller petal lengths and widths.
# 2.Versicolor lies in the middle of the other two in terms of petal length and width
# 3.Virginica has the largest of petal lengths and widths.
```

# Pair Plot

```python
sns.pairplot(dat,hue='Classification',height=2);

# Here, Iris Setosa has the smallest of petals widths and lengths. It also has the smallest sepal length but larger sepal widths.
```
