e. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this.

```python
# Calculate the math score cutoff for the top 25%
math_cutoff = st['math_score'].quantile(0.75)

# Identifying the students who are in the top 25% based on math score
top_25_percent_students = st[st['math_score'] >= math_cutoff]

# Assuming 'top_25_percent_students' is a slice of another DataFrame
top_25_percent_students = top_25_percent_students.copy()

# Giving bonus points to the top 25% students
# adding 5 bonus points
top_25_percent_students['math_score'] += 5

print("-----Top 25% students who have been awarded 5 bonus points------\n")
print(top_25_percent_students)
```

# 2. A company started to invest in digital marketing as a new way of their product promotions. For that they collected data and decided to carry out a study on it:

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
sale =pd.read_csv('/content/Sales_add.csv')
```

```python
sale.head()
```

a. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

```python
from scipy.stats import ttest_ind
```

```python
# Assigning the col values to a new variable
sales_before = sale['Sales_before_digital_add(in $)']
sales_after = sale['Sales_After_digital_add(in $)']

# Ho- Increase in sales after introducing digital marketing
# HA- There is no increase in sales after introduction of digital marketing
```

```python
# Perform two-tailed t-test
t_stat, p_value = ttest_ind(sales_before, sales_after)

print("Two tailed t-test Results:")
print("t-stat:", t_stat)
print("p-value:", p_value)


if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant increase in sales after starting digital marketing.")
else:
    print("Fail to reject the null hypothesis: There is no significant increase in sales after starting digital marketing.")
```

b. The company needs to check whether there is any dependency between the features “Region” and “Manager”

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(sale['Region'], sale['Manager'])

# creates a contigency table that  will have rows corresponding to the unique values of the 'Region' column and columns corresponding to the unique values of the 'Manager' column.
```

```python
# Perform chi-square test of independence
chi2_stat, p_value, dof, exp = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("There is a dependency between Region and Manager.")
else:
    print("There is no dependency between Region and Manager.")
```
